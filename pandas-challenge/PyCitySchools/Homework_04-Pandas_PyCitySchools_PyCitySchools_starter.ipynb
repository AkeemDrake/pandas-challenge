```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#districtsummary
students_count = school_data_complete["student_name"].count()
students_count
schools_count = len(school_data_complete["school_name"].unique())
schools_count
budget = school_data_complete["budget"].unique()
budget_total = budget.sum()
budget
avg_math = round(school_data_complete["math_score"].mean(), 6)
avg_math
avg_reading = round(school_data_complete["reading_score"].mean(), 6)
avg_reading
passing_score = round((avg_math + avg_reading)/2, 6)
passing_score
passing_math_df = school_data_complete.loc[school_data_complete['math_score'] >= 70]
passing_math = round((len(passing_math_df)/students_count)*100, 6)
passing_math
passing_reading_df = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
passing_reading = round((len(passing_reading_df) /students_count) *100, 6)
passing_reading
pd.options.display.float_format = '{:,}'.format
total_df = pd.DataFrame({"Total Schools": [schools_count], "Total Students": [students_count], "Total Budget": [budget_total], "Average Math Scores": [avg_math], 
"Average Passing Rate": [passing_score], "Students Passing Math": [passing_math], "Students Passing Reading": [passing_reading]})
total_df
total_df["Total Students"] = total_df["Total Students"].map("{:,}".format)
total_df["Total Budget"] = total_df["Total Budget"].map("${:,.2f}".format)
total_df
```

```python
#school summary
school_name_group = school_data_complete.groupby(["school_name", "type"])
school_count = school_name_group["student_name"].count()
avg_math_score = school_name_group["math_score"].mean()
avg_reading_score = school_name_group["reading_score"].mean()
budget = school_name_group["budget"].unique()
budget_per = budget/school_count
#passing_math = (school_name_group.loc[school_name_group['math_score'] >= 70])/ school_count * 100
avg_math = school_name_group["math_score"].mean()
avg_math
passing_math = avg_math.loc[avg_math >= 70] 
passing_math_group = passing_math_df.groupby(["school_name", "type"])
passing_reading_group = passing_reading_df.groupby(["school_name", "type"])
passing_math_group.count()
passing_math_percentage = passing_math_group['student_name'].count() / school_count * 100
passing_reading_percentage = passing_reading_group['student_name'].count() / school_count * 100
passing_rate = round((passing_math_percentage + passing_reading_percentage)/2, 6)
passing_rate

school_summary_df = pd.DataFrame({"Total Students": school_count, "Total School Budget": budget, "Per Student Budget": budget_per, "Average Math Score": avg_math_score, "Average Reading Score": avg_reading_score, "% Passing Math": passing_math_percentage, "% Passing Reading": passing_reading_percentage, "% Overall Passing Rate": passing_rate})
school_summary_df
```
